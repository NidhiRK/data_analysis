# relationship analysis¶

```python
df = data.drop(['Ticker'],axis = 1)
df.head()
```

```python
# Convert the 'Date' column to datetime data type
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Finally, to find the correlation among the variables, we can make use of the correlation function.
# This will give you a fair idea of the correlation strength between different variables.
corelation = df.corr()
```

```python
sns.heatmap(corelation,xticklabels = corelation.columns,yticklabels = corelation.columns,annot=True)
```

```python
sns.pairplot(df)
```

# data visualization

```python
 # Price Distribution
plt.hist(df['Open'], bins=20)  # histogram are used to represent the distribution of the data 
plt.title('market  Distribution')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(df.index,df['Close'], color='blue', linewidth=1)
plt.title('Stock Market Data - Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

```python
# 2. Bar Chart - Average Volume traded for each Ticker
avg_volume = data.groupby('Ticker')['Volume'].mean()
avg_volume.plot(kind='bar', figsize=(10, 6))
plt.title('Average Volume Traded by Ticker')
plt.xlabel('Ticker')
plt.ylabel('Average Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Box Plot - Spread of Close prices for each Ticker
data.boxplot(column='Close', by='Ticker', figsize=(10, 6))
plt.title('Spread of Close Prices by Ticker')
plt.xlabel('Ticker')
plt.ylabel('Close Price')
plt.show()
```

```python
# 5. Pie Chart - Proportion of Volume traded for each Ticker
total_volume = data.groupby('Ticker')['Volume'].sum()
total_volume.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%')
plt.title('Proportion of Volume Traded by Ticker')
plt.ylabel('')
plt.show()
```

```python
# 6. Scatter Plot - Relationship between High and Low prices
plt.scatter(data['High'], data['Low'])
plt.title('High vs Low Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.grid(True)
plt.show()
```
